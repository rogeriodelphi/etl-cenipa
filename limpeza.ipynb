```python
#Carregar o arquivo .csv que está separado por ";" excutar um parse na data e colocar o dia na primeira posição
df = pd.read_csv("ocorrencia_2010_2020.csv", sep=';', parse_dates=['ocorrencia_dia'], dayfirst=True)

#Traz os primeiros 5 registros
df.head()
```

```python
# Ver uma coluna que não tenha repetição de valores
df.codigo_ocorrencia. is_unique
```

```python
# # Mudar o "codigo_ocorrencia" para ser o novo índice
# "implace" atualiza a alteração no dataframe
df.set_index('codigo_ocorrencia', inplace=True)
```

```python
# listar os 5 primeiros registros
df.head()

# Buscar pelo novo índice
df.loc[40324]
```

```python
# Resetar a configuração de novo índece e voltando ao padrão
# "implace" atualiza a alteração no dataframe
df.reset_index(drop=True, inplace=True)

# listar os 5 primeiros registros
df.head()
```

```python
# Alterar um dado do campo "ocorrencia_aerodromo" para vazio
df.loc[0, 'ocorrencia_aerodromo'] = ''
df.head(1)
```

```python
#Alterar todo o conteúdo da linha 1 para 20
df.loc[1] = 20
df.head(2)
```

```python
# Alterar o valor coluna "total_recomendacoes" (linha a linha)
df.loc[:, 'total_recomendacoes'] = 0
df
```

```python
# Fazer o backup da coluna 'ocorrencia_uf' criando uma nova com nome 'ocorrencia_uf_bkp'
df['ocorrencia_uf_bkp'] = df.ocorrencia_uf
df
```

```python
# Apagar a coluna ocorrencia_uf_bkp
# axis=0 é para a linha e axis=1 é para coluna
# "implace" atualiza a alteração no dataframe
df.drop(['ocorrencia_uf_bkp'], axis=1, inplace=True)
df.head()
```

```python
# Onde a 'ocorrencia_uf' for igual a 'SP' altera a 'ocorrencia_classificacao' para 'GRAVE'
df.loc[df.ocorrencia_uf == 'SP', ['ocorrencia_classificacao']] = 'GRAVE'
df
```

```python
# listar todos os registros na qual 'ocorrencia_uf' for igual a 'SP'
df.loc[df.ocorrencia_uf == 'SP']
```

```python
df.head()
```

```python
# O 'ocorrencia_aerodromo' vai receber valor NA(não informado) onde o valor estiver '****'
df.loc[df.ocorrencia_aerodromo == '****', ['ocorrencia_aerodromo']] = pd.NA
df.head()
```

```python
# Vai alterar para NA(não informado) onde o valor estiver '**', '****',  '*****', '##', '####', 'NULL'
df.replace(['**', '****',  '*****', '###!', '####', 'NULL'], pd.NA, inplace=True)
df.head()
```

```python
# Quantos dados não informado eu tenho no meu DataFrame
# Somar por colunas quantos dados 'NA' nós temos
df.isna().sum()
# pode usar também a função 'df.isnull()'
```

```python
# Alterar 'NA' para o valor '10' 
# "implace" atualiza a alteração no dataframe
df.fillna(10, inplace=True)
df
```

```python
# Vai alterar para NA(não informado) onde o valor estiver '10' - Perigoso se não souber quais são os dados a alterar
# "implace" atualiza a alteração no dataframe
df.replace([10], pd.NA, inplace=True)
df.head()
```

```python
# Alterar para 10 na coluna 'total_recomendacoes' onde o valor era 'NA'
# "implace" atualiza a alteração no dataframe
df.fillna(value={'total_recomendacoes':10}, inplace=True)
df
```

```python
# apaga todas as linhas onde tinha 'NA'
df.dropna()
```

```python
# Apagar uma linha toda onde tiver 'NA' na coluna 'ocorrencia_uf' 
df.dropna(subset = ['ocorrencia_uf'])
```

```python
# Excluir valores duplicados 
df.drop_duplicates(inplace=True)
df
```
